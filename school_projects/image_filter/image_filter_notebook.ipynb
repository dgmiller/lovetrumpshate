# Image Filters
Using code written in the *image_filter.py* file, this notebook shows a brief application of applying filters to images when plotted as matrices. The specific filter used is called the *Sobel Filter* which applies a particular matrix to the image and finds areas of high contrast in the image.

First, we import the python file and matplotlib so we can plot the images.

```python
import image_filter
from matplotlib import pyplot as plt
%matplotlib inline
```

We will be working with the 'cameraman.png' file. Let's take a look at it in grayscale.

```python
pic = plt.imread('cameraman.png')
plt.imshow(pic, cmap='gray')
```

Next, we will run the function ***sobelFilter( )*** with the input image being 'cameraman.png'. The output is a new filtered image.

```python
filtered_pic = image_filter.sobelFilter(pic)
```

A side by side of the images is shown below for comparison. In our new image, the edges were detected where the change in contrast was greatest. This created a new matrix of zeros and ones as opposed to the matrix of our original image which was made of varied numbers.

```python
plt.subplot(1,2,1)
plt.imshow(filtered_pic, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(pic, cmap='gray')
```

```python
print "Filtered pic:\n\n", filtered_pic, "\n"
print "Original Matrix:\n\n", pic
```
